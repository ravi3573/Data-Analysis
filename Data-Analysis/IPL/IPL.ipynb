```python
teamss=(df_matches['team1'].value_counts()+df_matches['team2'].value_counts()).reset_index()
teamss.columns=['team_name','mtaches_played']
teamss
```

```python
df_matches['winner']=df_matches['winner'].map(Teams)
wins=df_matches['winner'].value_counts().reset_index()
wins.columns=['team_name','wins']
wins
```

```python
player=teamss.merge(wins,left_on='team_name',right_on='team_name',how='inner')
player
```

```python
player['%win']=(player['wins']/player['mtaches_played'])*100
player
```

```python
import plotly.offline as py
import plotly.graph_objs as go
```

```python
trace1=go.Bar(x=player['team_name'],y=player['mtaches_played'],name='Total_Matches')
trace2=go.Bar(x=player['team_name'],y=player['wins'],name='Matches_win')

data=[trace1,trace2]
py.iplot(data)
```

```python
df2=pd.read_csv(r'/kaggle/input/sports-dataset/deliveries/deliveries.csv')
df2.head()
```

```python
season=df_matches[['id','season']].merge(df2,left_on='id',right_on='match_id',how='left').drop('id',axis=1)
season
```

```python
season=season.groupby('season')['total_runs'].sum().reset_index()
season.set_index('season',inplace=True)
season.plot()
```

```python
avg_runs=df_matches.groupby('season')['id'].count().reset_index().rename(columns={'id':'matches'})
avg_runs
```

```python
season.reset_index(inplace=True)
```

```python
season=pd.concat([avg_runs,season.iloc[:,1]],axis=1)
season
```

```python
season['per_match_runs']=season['total_runs']/season['matches']
season.set_index('season')['per_match_runs'].plot()
```

```python
def comparison(team1,team2):
    compare=df_matches[((df_matches['team1']==team1)|(df_matches['team2']==team1))&((df_matches['team1']==team2)|(df_matches['team2']==team2))]
    sns.countplot(x='season',hue='winner',data=compare)
```
